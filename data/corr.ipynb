```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("trainset.csv", index_col="Region_Name")# dfr = dfr.T
dfr = df.T
dfr.rename(columns={"Region_Name":"Month"}, inplace=True)
dfr.insert(0, "Months", np.arange(1, 73))
dfr.reset_index(inplace=True)
dfr.drop(columns=["index"], inplace=True)
dfr
```

```python
for i in range(1000, 2000):#np.arange(1100, 1200):
    region = "Region " + str(i)
    if(dfr[region].max()-dfr[region].min()<10 and dfr[region].mean()>5 and dfr[region].mean()<10):
        dfr[region].plot(figsize=(20, 10), label = str(i))
        # plt.xlim(1, 74)
        print(i)
```
